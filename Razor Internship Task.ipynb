# Function to parse Ratings Value
A simple function which extracts only the rating value out of the whole string which was present in the original dataset

```python
def parse_ratings(string):
    x = string.split('%')
    return x[0]
```

# Function to parse Ratings Count
A simple function which extracts only the rating counts out of the whole string which was present in the original dataset

```python
def parse_ratings2(string):
    x = string.split("(")
    y = x[1].split(" ")
    return y[0]
```

```python
new_data['Rating Value'] = new_data['sellerratings'].apply(parse_ratings)
```

```python
new_data['Rating Counts'] = new_data['sellerratings'].apply(parse_ratings2)
```

```python
new_data.drop('sellerratings', axis=1, inplace = True)
```

# Function to parse Product Count
A simple function which extracts the product counts out of the whole string which was present in the original dataset

```python
def parse_product_count(string):
    if type(string) == str:
        x = string.split(" ")
        if len(x) == 5:
            return x[3]
        elif len(x) == 4:
            return x[2]
        else:
            return x[0]
    else:
        return 0
```

```python
new_data['Products Count'] = new_data['sellerproductcount'].apply(parse_product_count)
```

```python
new_data.drop('sellerproductcount', axis = 1, inplace = True)
```

# Function to parse Sellers Name
A simple function which extracts the seller's business name by removing the string literal "Business Name:" from it"

```python
def parse_seller_name(string):
    if type(string) == str:
        x = string.split(":")
        return x[1]
    else:
        return string
```

```python
new_data['Business Name'] = new_data['seller business name'].apply(parse_seller_name)
```

```python
new_data.drop('seller business name', axis = 1, inplace = True)
```

# Function to parse Address and Country
A function to parse out the Seller's Country which can be used to later filter other values out

```python
def parse_address(string):
    if type(string) == str:
        return string[-2:]
    else:
        return string
```

```python
new_data['Country'] = new_data['businessaddress'].apply(parse_address)
```

# Sorting Most Favourable Sellers

# Step 1. Filter all the sellers who have minimum 90 rating 
I have kept a minimum seller's rating threshold to be 90 because this will make sure we only engage with the best rated sellers out of all the possible ones. So the first primary filtering method is to filter down values based on a certain rating

```python
new_data = new_data[new_data['Rating Value'].astype(int) > 90]
```

# Step 2. Filter all the sellers who also have minimum 50 ratings
Now that we only have sellers who have a minimum rating of 90, in order to remove inconsistences I am filtering sellers who have got a minimum of 50 ratings on their products. This helps to ensure we have a credible average that is based on a decent number of users and not 1 or 2 reviews

```python
new_data = new_data[new_data['Rating Counts'].astype(int) > 50]
```

# Option 1. Sort the dataset according to Rating Value to find best sellers

Now I have taken 2 options. The first is to take a direct ordering of sellers according to their ratings and also the above mentioned filters. From this we can observe 3 things

1. The rating values are maximum and also have decent number of ratings as well 
2. The max % of negative reviews are 0 as they have achieved a perfect rating score over the last 12 months.
3. The countries which these sellers belong to is Denmark and its a majority.

This then is a very effective way to garner insights onto the most valuable/favourable sellers to engage with.

But the only drawback is that the Hero Product Rating counts of some sellers is very low and thus could be confidence breaking in terms of their credibility.

```python
new_data = new_data.sort_values(by = 'Rating Value')
```

```python
new_data.head()
```

# Option 2. Sort the dataset according to Hero Product Ratings

The second option involves sorting the sellers according to the total Hero Products rating counts. This involves the total number of reviews that the seller's most popular items have got. This will ensure a fairer ordering as it involves the most popular items of a seller which could lead to engagement solely based on customer engagement. 

The broad observation here is that this way, the dataset has been very balanced. There are visible negative reviews %, the ratings aren't all 100% and there are different countries who are now a part of the observation.

However a major disadvantage is that we no longer have a clear picture which sellers we can actively engage with. As all of them have fairly similar ratings, it becomes harder to narrow down or select the top 5 wihtout having to cross check other parameters


```python
new_data = new_data.sort_values(by = ['Hero Product 1 #ratings','Hero Product 2 #ratings'] , ascending = False)
```
